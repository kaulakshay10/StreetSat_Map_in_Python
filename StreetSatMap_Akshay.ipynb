```python
import pandas as pd
import os
from mapboxgl.utils import *
from mapboxgl.viz import *
```

Creating a dataframe consisting of latitudes and longitudes

```python
df=pd.DataFrame({'Lat':[37.774929],'Lon':[-122.419418]})
```

```python
df1=pd.DataFrame({'Lat':[34.052235],'Lon':[-118.243683]})
```

```python
df=df.append(df1,ignore_index=True)
```

Converting dataframe to *json* format


```python
dfjson= df_to_geojson(df,
                       #properties=['Example'],
                       lat='Lat', lon='Lon', precision=3)
```

Running the code to view the street satellite map

```python
for i in range(0,len(df)):
  viz = CircleViz(dfjson,
  access_token='pk.eyJ1Ijoia2F1bGFrc2hheTEwIiwiYSI6ImNrOTF3OWpseTAyNjUzaHA0dnV0YTVjY20ifQ.i9sIo7l8frMlSNdaDeFnFg',
  height='400px',
  stroke_color='black',
  center=(df.loc[i,'Lon'], df.loc[i,'Lat']),
  zoom=17,
  style='mapbox://styles/mapbox/satellite-streets-v11')
  print("Lat:{}  Lon:{}".format(df.loc[i,'Lat'],df.loc[i,'Lon']))
  viz.show()
```

Creating a function with latitude and longitude as input to display the street view map of the location

```python
def streetsatmap(lat,lon):
    if lat!=0 and lon!=0:
        df=pd.DataFrame([[lat,lon]])
        df.columns=['Lat','Lon']
        dfjson= df_to_geojson(df,
                       #properties=['Example'],
                       lat='Lat', lon='Lon', precision=3)
        viz = CircleViz(dfjson,
                access_token='pk.eyJ1Ijoia2F1bGFrc2hheTEwIiwiYSI6ImNrOTF3OWpseTAyNjUzaHA0dnV0YTVjY20ifQ.i9sIo7l8frMlSNdaDeFnFg',
                height='400px',
                stroke_color='black',
                center=(lon, lat),
                zoom=17,
                style='mapbox://styles/mapbox/satellite-streets-v11'
               )
        print("Satellite Street View for Lat:{} and Lon:{}".format(lat,lon))
        viz.show()
    else:
        print('Error: Parameter equals 0')
```
